### Q.1 What was the most common decision on winning the toss?

```python
decision=df["Toss Decision"]
print(decision)
print(decision.count())
print(type(decision))
```

```python
arr = decision.to_numpy()
print(arr)
```

```python
count = 0
for i in arr:
    if(i=="Field"):
         count=count+1
print("The number of times fielding was chosen is",count)
count1=decision.count()-count
print("The number of times batting was chosen is",count1)
import matplotlib.pyplot as plt
arr=numpy.array([count,count1])
mylabels=['Fielding','Batting']
plt.pie(arr,labels = mylabels)
plt.show()
if(count>count1):
    print("Fielding was the most common decision made by the team winning the toss.")
else:
    print("Batting was the most common decision made by the team winning the toss.")
```

### Q.3 Which team scored the highest runs in a match and in which Stadium?

```python
import numpy as np
Score_A = df['Score A']
Team_A = df['Team A']
Score_B = df['Score B']
Team_B = df['Team B']
Stadium = df['Stadium']
City = df['City']

Team_Aarr = Team_A.to_numpy()
A = np.array(Team_Aarr)
Team_Barr = Team_B.to_numpy()
B = np.array(Team_Barr)
ScoreArr1 = Score_A.to_numpy()
ScoreArr2 = Score_B.to_numpy()
Stadiumarr = Stadium.to_numpy()
Cityarr = City.to_numpy()

max_score_team_A = np.max(ScoreArr1)
max_score_team_B = np.max(ScoreArr2)

if max_score_team_A > max_score_team_B:
    indices_team_A = np.where(ScoreArr1 == max_score_team_A)[0]
    print(f"{A[indices_team_A]} scored the maximum score of {max_score_team_A} in the 2023 World Cup , at {Stadiumarr[indices_team_A]} in {Cityarr[indices_team_A]} ")
elif max_score_team_B > max_score_team_A:
    indices_team_B = np.where(ScoreArr2 == max_score_team_B)[0]
    print(f"{B[indices_team_B]} scored the maximum score of {max_score_team_B} in the 2023 World Cup , {Stadiumarr[indices_team_A]} in {Cityarr[indices_team_A]} ")
else:
    print(f"Both teams have the same maximum score.{Stadiumarr[indices_team_A]} in {Cityarr[indices_team_A]}")
```

### Q.4 Which player won the most Man of the Match awards in the tournament

```python
def most_occuring_element(arr):
    unique_elements, counts = np.unique(arr, return_counts=True)
    most_common_index = np.argmax(counts)
    most_common_element = unique_elements[most_common_index]
    return most_common_element
mostMOM=df.groupby(['Man of the Match']).size().reset_index(name='CountMOM')
print(mostMOM)
```

### Q.9. Plot a graph for Team v/s total runs in the tournament

```python
Team_A = df['Team A']
Team_B = df['Team B']
Team_Aarr = np.array(Team_A.to_numpy())
Team_Barr = np.array(Team_B.to_numpy())
Score_A = df['Score A']
Score_B = df['Score B']
ScoreArr1 = np.array(Score_A.to_numpy())
ScoreArr2 = np.array(Score_B.to_numpy())
print(Team_Aarr)
print(ScoreArr1)
dict = {}
j=0
for i in Team_Aarr:
    new_key = i
    if new_key in dict:
        ScoreArr1[j] = ScoreArr1[j]+dict[i]
        dict[i]= ScoreArr1[j] 
    else:
        dict[i]= ScoreArr1[j]
    j=j+1
print(dict)
j=0
for i in Team_Barr:
    new_key = i
    if new_key in dict:
        ScoreArr2[j] = ScoreArr2[j]+dict[i]
        dict[i]= ScoreArr2[j] 
    else:
        dict[i]= ScoreArr2[j]
    j=j+1
print(dict)
```

```python
import matplotlib.pyplot as plt
Countries = list(dict.keys())
TotalRuns = list(dict.values())
plt.figure(figsize=(8, 6))  
plt.bar(Countries, TotalRuns, color='skyblue') 
plt.xlabel('Team')
plt.ylabel('Total Runs')
plt.title('Bar Plot for Team V/s Total Runs')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
print("Hence, team India scored the most runs in World Cup 2023.")
```

### Q.10 Plot a graph of number of wides and no-balls by each team in the tournament.

```python
Team_A = df['Team A']
Team_B = df['Team B']
Team_Aarr = np.array(Team_A.to_numpy())
Team_Barr = np.array(Team_B.to_numpy())
Wides_A = df['Wides A']
Wides_Aarr1 = np.array(Wides_A.to_numpy())
Wides_B = df['Wides B']
Wides_Aarr2 = np.array(Wides_B.to_numpy())
NoBalls_A = df['No Balls A']
NoBalls_Aarr1 = np.array(NoBalls_A.to_numpy())
NoBalls_B = df['No Balls B']
NoBalls_Aarr2 = np.array(NoBalls_B.to_numpy())
dict2={}
j=0
for i in Team_Aarr:
    new_key = i
    if new_key in dict2:
        temp = Wides_Aarr1[j]+NoBalls_Aarr1[j]
        temp = temp + dict2[i]
        dict2[i]= temp
    else:
        dict2[i]= Wides_Aarr1[j]+NoBalls_Aarr1[j]
    j=j+1
print(dict2)
j=0
for i in Team_Barr:
    new_key = i
    if new_key in dict2:
        temp = Wides_Aarr2[j]+NoBalls_Aarr2[j]
        temp = temp + dict2[i]
        dict2[i]= temp
    else:
        dict2[i]= Wides_Aarr2[j]+NoBalls_Aarr2[j]
    j=j+1
print(dict2)
```

```python
import matplotlib.pyplot as plt
Countries = list(dict2.keys())
WN = list(dict2.values())
plt.figure(figsize=(8, 6))  
plt.bar(Countries, WN, color='skyblue') 
plt.xlabel('Team')
plt.ylabel('Total Wides and No-balls')
plt.title('Bar Plot for Team V/s Total Wides and No-balls')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
print("Hence , South Africa gave most extras in the World Cup 2023")
```
